```python
import os
import pandas as pd

# Specify the folder containing CSV files
folder_path = '/content/cfi/'

# Get a list of all CSV files in the folder
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# Initialize an empty DataFrame to store the concatenated data
concatenated_data = pd.DataFrame()

# Loop through each CSV file and concatenate the data
for csv_file in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    data = pd.read_csv(file_path)
    concatenated_data = pd.concat([concatenated_data, data], ignore_index=True)

# Display the shape of the concatenated DataFrame
print(f'Shape of concatenated data: {concatenated_data.shape}')

# Optionally, save the concatenated DataFrame to a new CSV file
concatenated_data.to_csv('/content/concatenated_data.csv', index=False)
```

```python
import random
nrow = random.randint(0, concatenated_data.shape[0])
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_images(data, nimages, rows=5):
    values = data.iloc[:, 12:].values
    values = np.array(values).reshape(-1, 20, 20)

    cols = (nimages + rows - 4) // rows
    plt.figure(figsize=(10, 8))

    for i in range(nimages):
        plt.subplot(rows, cols, i + 1)
        plt.imshow(values[i], cmap='gray')
        plt.title(f'{data.iloc[i, 2]} - {data.iloc[i, 0]} - {chr(data.iloc[i, 2]).encode("utf-8").decode("utf-8")}')
        plt.axis('off')

    plt.show()

def main():
    nrows = 25
    data = pd.read_csv('concatenated_data.csv', skiprows=range(1, nrow - nrows), nrows=nrows)

    print(nrow)
    plot_images(data, nrows)

if __name__ == "__main__":
    main()
```

```python
def plot_categories(training_images, training_labels):
    fig, axes = plt.subplots(5, 5, figsize=(16, 8))
    axes = axes.flatten()

    for k in range(nrows):
        img = training_images[k]
#         img = np.expand_dims(img, axis=-1)
#         img = array_to_img(img)
        ax = axes[k]
        ax.imshow(img, cmap="Greys_r")
        ax.set_title(f'{chr(training_labels[k]).encode("utf-8").decode("utf-8")}')
        # ax.set_title(f"{chr(int(training_labels[k])+97)}")
        ax.set_axis_off()

    plt.tight_layout()
    plt.show()

plot_categories(images, labels)
```
