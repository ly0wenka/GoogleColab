Import Libraries

```python
import numpy as np
from minisom import MiniSom
import matplotlib.pyplot as plt
```

Task. Build a neural network with a Kohonen layer, which will divide the
set of input data into clusters and define their centers, using the built-in
functions of the neural networks package of the MATLAB mathematical
environment. Add a new input vector to the trained network and determine
the cluster to which it belongs.
To create a neural network with a Kohonen layer, we will use the built-in
MATLAB environment function newc:

```python
# Setting the range of elements change
X = np.array([[0, 1], [0, 1]])

# Setting the number of clusters, points, and standard deviation
clusters = 5
points = 5
std_dev = 0.01

# Modeling of the input data
def nngenc(X, clusters, points, std_dev):
    np.random.seed(42)  # for reproducibility
    P = []
    for _ in range(clusters):
        center = np.random.rand(2, 1)
        cluster_points = center + std_dev * np.random.randn(2, points)
        P.append(cluster_points)
    return np.concatenate(P, axis=1)

P = nngenc(X, clusters, points, std_dev)

# Creating a Kohonen layer
som_size = (5, 5)  # Set the grid size
h = MiniSom(som_size[0], som_size[1], P.shape[0], sigma=1.0, learning_rate=0.1)

# Setting the number of training cycles
epochs = 50
h.train_random(P.T, epochs)

# Get the weight vectors from the trained SOM
w = h.get_weights()

# Output of graphs of the input data and identified cluster centers
plt.plot(P[0, :], P[1, :], '^r')
plt.grid()
plt.plot(w[:, :, 0], w[:, :, 1], 'ob')
plt.xlabel('p(1)')
plt.ylabel('p(2)')
plt.show()

# Setting a new input vector
A = 0.6
B = 0.5
p = np.array([A, B])

# Network scanning
winner = h.winner(p)
plt.plot(P[0, :], P[1, :], '^r')
plt.grid()
plt.plot(w[:, :, 0], w[:, :, 1], 'ob')
plt.plot(A, B, '+k')
plt.plot(w[winner[0], winner[1], 0], w[winner[0], winner[1], 1], 'sg', markersize=10)  # highlight the winning neuron
plt.xlabel('p(1)')
plt.ylabel('p(2)')
plt.show()
```

```python
from numpy import genfromtxt

# Load the clusterdemo.dat file
clusterdemo = genfromtxt('clusterDemo.csv', delimiter=',')
```

```python
# Extract the first two columns and transpose
P = clusterdemo[:, :2].T * 100

# Creating a Kohonen NN with 9 clusters
som_size = (3, 3)  # Set the grid size
h = MiniSom(som_size[0], som_size[1], P.shape[0], sigma=1.0, learning_rate=0.1)

# Setting the number of training cycles
epochs = 100
h.train_random(P.T, epochs)

# Get the weight vectors from the trained SOM
w = h.get_weights()

# Output of graphs of the input data and identified cluster centers
plt.plot(P[0, :], P[1, :], '^r')
plt.grid(True)
plt.hold(True)
plt.plot(w[:, :, 0], w[:, :, 1], 'ob')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()

# Setting a new input vector
A = 62
B = 50
p = np.array([A, B])

# Network scanning
winner = h.winner(p)
plt.plot(P[0, :], P[1, :], '^r')
plt.grid(True)
plt.hold(True)
plt.plot(w[:, :, 0], w[:, :, 1], 'ob')
plt.plot(A, B, '+k')
plt.plot(w[winner[0], winner[1], 0], w[winner[0], winner[1], 1], 'sg', markersize=10)  # highlight the winning neuron
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()

print(f"The new input vector belongs to cluster {winner}")
```
